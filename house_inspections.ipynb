```python
# Set up the map.
centre = [-37.77914103182415, 144.9783582972783]
m = Map(center=centre, zoom=13)
m.layout.height = '600px'

# Create a polygon with a shaded fill.
region = Polygon(
    locations=boundary,
    color="blue",        # Outline color
    fill_color="blue",   # Fill color
    fill_opacity=0.2     # Adjust opacity as needed (0 is transparent, 1 is solid)
)

# Add the polygon to your map (assuming m is your Map object)
m.add_layer(region)


# Create an HTML widget to display marker information.
info = widgets.HTML(
    value="<h3>Click on a marker to see more information</h3>",
    layout=widgets.Layout(width='100%', height='150px', overflow_y='auto')
)
display(info)

# Callback for house markers that displays detailed info including an image.
def make_house_callback(house_info):
    def callback(event=None, **kwargs):
        text = f"""
        <div style="display: flex; align-items: flex-start;">
            <div style="flex: 1;">
                <b>Address:</b> {house_info['address']}<br>
                <b>Bedrooms:</b> {house_info['bedrooms']}<br>
                <b>Price:</b> {house_info['price']}<br>
                <b>Inspection Date:</b> {house_info['inspection_date']}<br>
                <b>Inspection Time:</b> {house_info['inspection_time']}<br>
                <a href="{house_info['listing_url']}" target="_blank">Listing URL</a>
            </div>
            <div style="margin-left: 20px;">
                <img src="{house_info['image_url']}" alt="House image" style="max-width:200px;"/>
            </div>
        </div>
        """
        info.value = text
    return callback

def make_place_callback(place_info):
    def callback(event=None, **kwargs):
        # Set the font size to 24px (or any size you prefer)
        info.value = f"<div style='font-size: 24px;'><b>{place_info['name']}</b></div>"
    return callback

# First add the places markers with a lower z-index.
for place in places:
    icon = AwesomeIcon(
        name='info-sign',
        marker_color='green',
        icon_color='white'
    )
    marker = Marker(location=place['location'], icon=icon, draggable=False)
    marker.z_index_offset = 500  # Lower z-index so these markers are drawn beneath houses if they overlap.
    marker.on_click(make_place_callback(place))
    m.add_layer(marker)

# Then add the houses markers with a higher z-index.
for house in houses:
    marker = Marker(location=house['coords'], draggable=False)
    marker.z_index_offset = 1000  # Higher z-index to ensure these markers appear on top.
    marker.on_click(make_house_callback(house))
    m.add_layer(marker)

# Display the final map.
display(m)
```
